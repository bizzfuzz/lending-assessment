# Data Analysis

```python
import pandas as pd
import numpy as np
from pandas import to_timedelta


disbursements = pd.read_pickle("exports/disbursements_clean.pkl")
repayments = pd.read_pickle("exports/repayments_accounts.pkl")
```

```python
disbursements["due_date"] = disbursements.date + to_timedelta(disbursements.tenure, unit="D")
disbursements
```

```python
#group repayments by account and total amount
agg_repayments = repayments.groupby("account").agg({"amount": "sum"}).reset_index()
agg_repayments
```

Matching loans with their repayments

```python
loan_repayments = pd.merge(disbursements, agg_repayments, left_on="account", right_on="account", how="left")
loan_repayments.rename(columns={"amount_x": "disbursed_amount", "amount_y": "repaid_amount", "due":"amount_due"}, inplace=True)
loan_repayments
```

Assigning the latest payment date to each loan

```python
last_payment_date = repayments.groupby("account").agg({"date": "max"}).reset_index()
loan_repayments = pd.merge(loan_repayments, last_payment_date, left_on="account", right_on="account", how="left")
loan_repayments.rename(columns={"date_x": "disbursement_date", "date_y" : "last_payment_date"}, inplace=True)
loan_repayments
```

Engineering fully repaid column

```python
loan_repayments["fully_repaid"] = loan_repayments.repaid_amount >= loan_repayments.disbursed_amount
loan_repayments
```

Engineering outstanding amount column. Some loans overpaid, so limited to 0

```python
loan_repayments["outstanding_amount"] = loan_repayments.disbursed_amount - loan_repayments.repaid_amount
loan_repayments["outstanding_amount"] = loan_repayments["outstanding_amount"].clip(lower=0)
loan_repayments
```

```python
loan_repayments["repaid_late"] = loan_repayments.last_payment_date > loan_repayments.due_date
loan_repayments
```

### Engineering columns for easier analysis

```python
loan_repayments["effective_interest_rate"] = (loan_repayments["fee"] / loan_repayments["disbursed_amount"]) * (365 / loan_repayments["tenure"])
loan_repayments
```

```python
num_payments = repayments.groupby("account")["date"].count().reset_index(name="num_payments")
loan_repayments = loan_repayments.merge(num_payments, on="account", how="left")
loan_repayments
```

```python
loan_repayments["days_late"] = (loan_repayments["last_payment_date"] - loan_repayments["due_date"]).dt.days.clip(lower=0)
loan_repayments
```

```python
latest_date = loan_repayments["disbursement_date"].max()
loan_repayments["overdue"] = (loan_repayments["due_date"] < latest_date) & (~loan_repayments["fully_repaid"])
loan_repayments
```

```python
loan_repayments["active"] = (~loan_repayments.fully_repaid) & (~loan_repayments.overdue)
loan_repayments
```

```python
conditions = [
    loan_repayments["active"],
    loan_repayments["overdue"],
    loan_repayments["fully_repaid"]
]
choices = ["active", "overdue", "repaid"]

loan_repayments["status"] = np.select(conditions, choices, default="unknown")
loan_repayments
```

```python
loan_repayments.to_pickle("exports/loan_repayments.pkl")
loan_repayments.to_csv("exports/loan_repayments.csv", index=False)
```

```python
monthly_summary =loan_repayments.copy()
monthly_summary["month"] = monthly_summary["disbursement_date"].dt.to_period("M")
monthly_summary = monthly_summary.groupby("month")[["disbursed_amount", "repaid_amount", "outstanding_amount", "fee"]].sum()
monthly_summary
```

```python
monthly_summary.index = monthly_summary.index.to_timestamp()
monthly_summary['revenue_change'] = monthly_summary['repaid_amount'].pct_change()
monthly_summary['disbursement_change'] = monthly_summary['disbursed_amount'].pct_change()
monthly_summary = monthly_summary.dropna()
monthly_summary = monthly_summary.loc[monthly_summary.index.month != 8]
monthly_summary
```

Forcating disbursements and repayments for the following 3 months. July used as a bas as August data is not for the whole month.

```python
last_revenue = monthly_summary['repaid_amount'].iloc[-1]
avg_growth_rate = monthly_summary['revenue_change'].mean()

# Forecast next 3 months
forecast_months = 3
forecast = [last_revenue * ((1 + avg_growth_rate) ** i) for i in range(1, forecast_months + 1)]
#round all forecasts to decimal places
forecast = [round(i, 2) for i in forecast]
forecast
```

```python
last_disbursement = monthly_summary["disbursed_amount"].iloc[-1]
avg_growth_rate = monthly_summary["disbursement_change"].mean()
disbursement_forecast = [last_disbursement * ((1 + avg_growth_rate) ** i) for i in range(1, forecast_months+1)]
disbursement_forecast = [round(i,2) for i in disbursement_forecast]
disbursement_forecast
```

```python
future_months = pd.date_range(start=monthly_summary.index[-1] + pd.DateOffset(months=1), periods=forecast_months, freq='M')
forecast_df = pd.DataFrame({'month': future_months, 'repaid_amount': forecast, 'disbursed_amount' : disbursement_forecast})
monthly_summary = pd.concat([monthly_summary, forecast_df.set_index('month')])
monthly_summary
```

```python
monthly_summary.to_csv("exports/monthly_summary.csv")
```
